## Time Resampling

```python
df_nvda.head()
```

```python
df_nvda.resample(rule='A').min()
```

```python
df_nvda.resample(rule='A').max()
```

```python
##year end frequency
df_nvda.resample(rule='A').max()['Open'].plot()
```

```python
##quaterly start frequency
##https://towardsdatascience.com/resample-function-of-pandas-79b17ec82a78
df_nvda.resample(rule='QS').max()['High'].plot()
```

```python
##Business End Frequency
##https://towardsdatascience.com/resample-function-of-pandas-79b17ec82a78
df_nvda.resample(rule='BA').max()
```

```python
df_nvda.resample(rule='BQS').max()
```

```python
##plotting
df_nvda['Open'].resample(rule='A').mean().plot(kind='bar')
```

```python
df_nvda['Open'].resample(rule='M').max().plot(kind='bar',figsize=(15,6))
```

```python
df_nvda['High'].rolling(11).max().head(20)
```

```python
df_nvda.head()
```

```python
df_nvda['Open:30 days rolling']=df_nvda['Open'].rolling(30).mean()
```
